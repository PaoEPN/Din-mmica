Planteamiento de ecuaciones<br>
Sumatoria de fuerzas en el eje x<br>
$\sum \vec{F_{x}}=0$<br>$Nsen(\Theta )-Frcos(\Theta )=0$<br>$Nsen(\Theta )-N\mu cos(\Theta )=0$<br>
$N(sen(\Theta )-\mu cos(\Theta ))=0$<br>
$N=0$<br>
$sen(\Theta )-\mu cos(\Theta )=0$<br>
$sen(\Theta )=\mu cos(\Theta )$<br>
$tan(\Theta )=\mu $<br>
$\Theta =16.70° = 0.29 rad$<br>

Sumatoria de fuerzas en el eje y<br>
$\sum \vec{F_{y}}=0$<br>$Ncos(\Theta )+Frsen(\Theta )-mg=0$<br>
$Ncos(\Theta )+N\mu sen(\Theta )-mg=0$<br>
$N(cos(\Theta )+\mu sen(\Theta ))=mg$<br>
$N=\frac{mg}{cos(\Theta )+\mu sen(\Theta )}$<br>
Se grafica la ecuación

```python
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go

#Ejercico que no se grabó

#datos
m=2000
g=9.81
u=0.3

R = 16/1000 #[m]
O=2000*((2*np.pi)/60) #[rad/s]
#análisis
th=np.arange(0,17,0.1) #creando lista numérica
vth = np.array(np.deg2rad(th)) #transformado a una matriz
vs=m*g/(u*np.sin(vth)+np.cos(vth))
#plot
fig=px.line(x=vth,y=vs,labels={'x':'Theta[rad]','y':'Normal[N]'},title="Normal - ángulo theta")
fig.show()
```

**Literal b**<br>
Para velocidad angular <br>
Sumatoria de torques en G<br>
$\sum\vec{ \tau _{O}}=\frac{d}{dt}\vec{H_{O}}+\vec{v_{G/O}} x \vec{p_{G/O}}=\frac{d}{dt}I_{zz}\omega _{z} \vec{k}$<br>
$\vec{v_{G/O}} x \vec{p_{G/O}}=0$<br><br>
Para calcular la inercia de la placa se utiliza el teorema de los ejes paralelos<br>
Ubicando al eje en el centro de gravedad G la inercia es <br>
$I_{CM}=\frac{1}{12}m(a^{2}+b^{2})$<br>
$a=b$<br>
$I_{CM}=\frac{1}{6}mb^{2}$<br><br>
Aplicando el teorema de los ejes paralelos<br>


$I=I_{CM}+mD^{2}$<br>
$D=b/2$<br>
$I_{zz}=\frac{1}{6}mb^{2}+m\frac{b}{2}^{2}$<br>
$I_{zz}=\frac{1}{6}mb^{2}+\frac{1}{4}mb^{2}$<br>
$I_{zz}=\frac{5}{12}mb^{2}$<br>

Entonces:<br>
$\vec{H_{O}}=\frac{5}{12}mb^{2}\omega _{z}\vec{k}$<br>
$\sum \vec{\tau _{O}}=\frac{d}{dt}\vec{H_{O}}=\frac{5}{12}mb^{2}\ddot{\theta }\vec{k}$<br><br>

Haciendo sumatoria de torques en O<br>
$\sum \vec{\tau _{O}}=mgsen(\theta )\frac{b}{2}\vec{k}$<br>

Igualando<br>
$\frac{5}{12}mb^{2}\ddot{\theta }\vec{k}=mgsen(\theta )\frac{b}{2}\vec{k}$<br>
$\ddot{\theta }=1.128\frac{rad}{s^{2}}\vec{k}$<br>
$2\ddot{\theta }\theta=\dot{\Theta }_{f}^{2}-\dot{\Theta }_{o}^{2}$<br>
$\dot{\Theta }_{o}=0$<br>
$\dot{\Theta }_{f}=\sqrt{2\ddot{\Theta }\Theta } $<br>
$\dot{\Theta }_{f}=0.81 \frac{rad}{s}\vec{k} $

Ahora graficando<br>
$\dot\Theta=\sqrt{2\ddot{\Theta }\Theta } $<br>

```python
#análisis
t2=1.128
th=np.arange(0,17,0.1) #creando lista numérica
vth = np.array(np.deg2rad(th)) #transformado a una matriz
vs=np.sqrt(2*t2*vth)
#plot
fig=px.line(x=vth,y=vs,labels={'x':'Theta[rad]','y':'Velocidad angular[rad/s]'},title="Velocidad angular - ángulo theta")
fig.show()
```
